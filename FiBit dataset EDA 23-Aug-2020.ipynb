```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#read data
fit_data = pd.read_csv("C:/Users/RAAVAN/Downloads/Python/FitBit data.csv")
fit_data.head()
```

```python
#Data information
fit_data.info()
```

ActivityDate is the only object type data.

```python
fit_data.describe()
```

```python
#relationship between two variables TotalSteps & Calories
sns.set(style="whitegrid")
plt.figure(figsize=(8, 8))

sns.regplot(x= 'TotalSteps', y= 'Calories', data= fit_data)
```

This plot tell us that TotalSteps & Calories has positive relationship. That mean step increases and calorie burn also increases.

```python
#Make a new column Daily_ActiveMin_total 
fit_data['Daily_ActiveMin_total'] = fit_data[['VeryActiveMinutes', 'FairlyActiveMinutes', 'LightlyActiveMinutes']].sum(axis=1)
fit_data.head()
```

```python
#relationship between two variables Daily_ActiveMin_total & Calories
sns.set(style="whitegrid")
plt.figure(figsize=(8, 8))

sns.regplot(x= 'Daily_ActiveMin_total', y= 'Calories', data= fit_data)
```

Calories burning increase with Daily_ActiveMin_total increases 

```python
#Person avg. daily activity & sedentary times
fit_data['Daily_ActiveMin_total'].mean(), fit_data['SedentaryMinutes'].mean()
```

Person avg. daily activity near 200 min where it is included jogging, walking, go for market, office.
Person avg. daily sits  995 min including sleeping time also.

```python
#Find out average speed of person
fit_data['speed_km'] = (fit_data['TotalDistance'])/(fit_data['Daily_ActiveMin_total']/60)
fit_data['speed_km'].mean()
```

mean value is infinitive for it because Daily_ActiveMin_total in many rows are zero

```python
#Find out the Abnormality
fit_data[(fit_data.Daily_ActiveMin_total==0) & (fit_data.TotalSteps!= 0)]
```

Person total steps more than zero even tracker distance shows moore than zero km but Daily_ActiveMin_total is zero

```python
# Drop rows where Daily_ActiveMin_total ==0
new_fit_data=fit_data[(fit_data.Daily_ActiveMin_total > 0)]
new_fit_data.info()
```

```python
new_fit_data['speed_km'] = (new_fit_data['TotalDistance'])/(new_fit_data['Daily_ActiveMin_total']/60)
new_fit_data['speed_km'].mean()
```
